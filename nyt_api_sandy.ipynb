# API Call to NYT

```python
url = f'https://api.nytimes.com/svc/search/v2/articlesearch.json?q=Hurricane Sandy&api-key={nyt_key}'
response = requests.get(url).json()
```

```python
contents = response['response']['docs']
#pprint(contents)
```

```python
news_list = []
for article in contents:
    news_dict={}
    news_dict['Headlines'] = (article['headline']['main'])
    news_dict['Contents'] = (article['lead_paragraph'])
    news_dict['Source'] = (article['source'])
    news_dict['Pub Date'] = (article['pub_date'])
    news_dict['URL:'] = (article['web_url'])
    news_list.append(news_dict)
```

```python
news_df = pd.DataFrame(news_list)
news_df
```

```python
url2 = f'https://api.nytimes.com/svc/archive/v1/2012/10.json?api-key={nyt_key}'
response2 = requests.get(url2).json()
content = response2['response']['docs']
```

```python
content_df = pd.DataFrame(content)
#content_df.head()
```

```python
narrow_content_df = content_df.loc[content_df['news_desk']=='New York']
narrow_content_df = narrow_content_df.loc[narrow_content_df['pub_date']>='2012-10-28T18:05:43+0000']
```

```python
final_archive_df = narrow_content_df[['abstract','lead_paragraph','source','pub_date','web_url']]
final_archive_df = final_archive_df.rename(columns={'abstract':'Headlines','lead_paragraph':'Contents',
                                                    'source':'Source','pub_date':'Pub Date',
                                                    'web_url':'URL:'})
final_archive_df
```

# Merge News DataFrames 

```python
final_news_df = pd.concat([news_df,final_archive_df])
final_news_df = final_news_df.reset_index(drop=True)
final_news_df
```
